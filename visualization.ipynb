```python
import json

f = open('4-bash_files_list_sc.json')

data = json.load(f)
```

```python
bash_files_list = data["bash_files"]
print(len(bash_files_list))
```

```python
num_lines_list = []
num_func_list = []
num_fors_list = []
num_ifs_list = []
for bash_file in bash_files_list:
    num_lines_list.append(int(bash_file["num_lines"]))
    num_func_list.append(int(bash_file["num_func"]))
    num_fors_list.append(int(bash_file["num_for"]))
    num_ifs_list.append(int(bash_file["num_if"]))
```

```python
print(num_lines_list)
```

```python
def countGreaterThan(test_list, k):
    count = 0
    for i in test_list:
        if i > k:
            count = count + 1
    return count
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import *

# binwidth = 5000
plt.hist(num_lines_list, bins=np.arange(0, 800, 5))
plt.ylim(ymin=0, ymax = 700)
plt.title("Number of Lines vs Count")

print(countGreaterThan(num_lines_list, 800))

```

```python
plt.hist(num_func_list, bins=np.arange(0, 30, 1))
plt.ylim(ymin=0, ymax = 5100)
plt.title("Number of functions vs Count")

print(countGreaterThan(num_func_list, 30))
```

```python
plt.hist(num_fors_list, bins=np.arange(0, 25, 1))
plt.ylim(ymin=0, ymax = 5100)
plt.title("Number of fors vs Count")

print(countGreaterThan(num_fors_list, 25))
```

```python
plt.hist(num_ifs_list, bins=np.arange(0, 40, 1))
plt.ylim(ymin=0, ymax = 3000)
plt.title("Number of ifs vs Count")

print(countGreaterThan(num_ifs_list, 40))
```

## Average numbers ##

```python
avg_lines = float(sum(num_lines_list)) / float(len(bash_files_list))
avg_funcs = float(sum(num_func_list)) / float(len(bash_files_list))
avg_fors = float(sum(num_fors_list)) / float(len(bash_files_list))
avg_ifs = float(sum(num_ifs_list)) / float(len(bash_files_list))
```

```python
print("Average number of lines: " + str(avg_lines))
print("Average number of functions: " + str(avg_funcs))
print("Average number of fors: " + str(avg_fors))
print("Average number of ifs: " + str(avg_ifs))
```

## Median Numbers ##

```python
import statistics

median_lines = statistics.median(num_lines_list)
median_func = statistics.median(num_func_list)
median_fors = statistics.median(num_fors_list)
median_ifs = statistics.median(num_ifs_list)
```

```python
print("Median number of lines: " + str(median_lines))
print("Median number of functions: " + str(median_func))
print("Median number of fors: " + str(median_fors))
print("Median number of ifs: " + str(median_ifs))
```
